```python
import sys
import numpy as np
import corner

chainfile = "nled_phy_SN+RiessH0+BBAO_mh_2000_1.txt"
ndim = 5
skip = 1000
```

```python
npchain = np.loadtxt(chainfile)
samples = npchain[skip:,2:ndim+2]
weights = npchain[skip:,0]

rootname = chainfile.strip("_1.txt").strip(".txt")

parsfile = rootname+'.paramnames'
print(parsfile)

parnames = open(parsfile,'r')

latexnames=[]
for row in parnames:
	latexnames.append(row[0])
parnames.close()
```

```python
figure = corner.corner(samples, labels=latexnames[0:ndim],\
                       bins = 20,\
                       weights = weights,\
                       color='g',\
                       quantiles=[0.5],\
                       show_titles=True,\
                       title_fmt = '.4f',\
                       smooth=True,\
                       smooth1d=True,\
                       fill_contours=True,\
                       title_kwargs={"fontsize": 12})

#figure.savefig(parsfile+"burn.png")
#figure.savefig(parsfile+"burn.pdf")
```
