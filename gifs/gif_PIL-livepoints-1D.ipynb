```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from time import sleep
import glob
import re
import os
wd = os.getcwd()# Figures out the absolute path for you in case your working directory moves around.
outputs_dir = wd+'/img/'
outputs_dir
```

```python
def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    '''
    alist.sort(key=natural_keys) sorts in human order
    http://nedbatchelder.com/blog/200712/human_sorting.html
    (See Toothy's implementation in the comments)
    '''
    return [ atoi(c) for c in re.split(r'(\d+)', text) ]
```

```python
chain = np.loadtxt('../chains/waCDM_phy_SN+HD+fs8+Planck_nested_dynesty_multi_1.txt', usecols=range(2,2+6))
np.shape(chain)
```

```python
fig = plt.figure()

label = 1
lp = 50
it = 0
    
for i in range(1900):
    if i>= 50 and i%50 == 0:
        it += 1
        chain_aux = chain[:it*lp, :]
        print(lp, i)
        plt.figure(figsize=(10, 10))
        plt.plot(chain_aux[-lp:, 0], marker='o', linestyle = 'None')
        plt.xlabel("$\\Omega_m$", fontsize=20)
        plt.ylabel("Value", fontsize=20)
#         plt.xlim(0.05, 0.5)
        plt.ylim(0.05, 0.5)
#         plt.clf()
        plt.savefig(outputs_dir+'image_{}.png'.format(label))
        print('saving {}image_{}.png'.format(outputs_dir, label))
        label += 1
```
