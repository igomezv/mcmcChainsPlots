```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

#samples_file = np.loadtxt('owaCDM_BBAO+SN+RiessH0_snest_0.5.txt', delimiter=" ")
#samples_file = np.loadtxt('owaCDM_BBAO+SN+RiessH0_mnest_0.5.txt', delimiter=" ")
#samples_file = np.loadtxt('owaCDM_BBAO+SN+RiessH0_bambi_0.5.txt', delimiter=" ")
#samples_file = np.loadtxt('owaCDM_BBAO+SN+RiessH0_sbambi_0.5.txt', delimiter=" ")
samples_file = np.loadtxt('owaCDM_phy_BBAO+SN+RiessH0_mh_2000_1.txt')

```

```python
#fig = plt.figure()
#fig, axs = plt.subplots(5,5, constrained_layout=True)
fig, axs = plt.subplots(3,3)

#ax = fig.add_subplot(111, projection='3d')
#Cambiar el numero de filas (nsamples) para variar muestreos a graficar
nsamples = 1500
samples_selection = samples_file[0:nsamples,:]
nsamplesstr = str(nsamples)

om = samples_selection[:,2]
obh2 = samples_selection[:,3]
h = samples_selection[:,4]
w = samples_selection[:,5]
wa = samples_selection[:,6]
ok = samples_selection[:,7]


axs[0,0].scatter(om, obh2, c='g', marker='.')
axs[0,0].set(xlabel='$\Omega_m$', ylabel='$\Omega_bh^2$')
axs[0,1].scatter(om, h, c='r', marker='.')
axs[0,1].set(xlabel='$\Omega_m$', ylabel='$h$')
axs[0,2].scatter(om, w, c='b', marker='.')
axs[0,2].set(xlabel='$\Omega_m$', ylabel='$\omega$')

axs[1,0].scatter(om, wa, c='m', marker='.')
axs[1,0].set(xlabel='$\Omega_m$', ylabel='$\omega_a$')
axs[1,1].scatter(om, ok, c='k', marker='.')
axs[1,1].set(xlabel='$\Omega_m$', ylabel='$\Omega_k$')


#fig.delaxes(axs[1,2])
axs[1,2].scatter(obh2, h, c='r', marker='.')
axs[1,2].set(xlabel='$\Omega_bh^2$', ylabel='$h$')

axs[2,0].scatter(om, w, c='b', marker='.')
axs[2,0].set(xlabel='$\Omega_bh^2$', ylabel='$\omega$')

axs[2,1].scatter(om, wa, c='m', marker='.')
axs[2,1].set(xlabel='$\Omega_bh^2$', ylabel='$\omega_a$')
axs[2,2].scatter(om, ok, c='g', marker='.')
axs[2,2].set(xlabel='$\Omega_bh^2$', ylabel='$\Omega_k$')


#fig.delaxes(axs[2,0])
#fig.delaxes(axs[2,1])

fig.tight_layout()
plt.subplots_adjust(top=2, bottom=1, left=0, right=1, hspace=0.5,
                    wspace=1)

plt.suptitle(nsamplesstr +" samples with Metropolis-Hastings", x=0.5, y=2.15, fontsize=20, color= 'gray')

plt.savefig(nsamplesstr + "_owacdm_mh.png", bbox_inches='tight')
```
