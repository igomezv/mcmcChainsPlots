**¿Por qué la evidencia bayesiana?**

La distribución posterior de los parámetros $\theta$ está dada por:

$P(\theta | D, H_i) = \frac{P(D|\theta,H_i) P(\theta|H_i)}{P(D|H_i)}$

donde:

$Evidencia = P(D|H_i) = \int P(D|\theta, H_i) P(\theta|H_i) d\theta$

- BIC, AIC, etc son aproximaciones.

- La evidencia bayesiana propone una opción rigurosa (y aceptada por la comunidad) para realizar la comparación.


- **La razón entre las evidencias bayesianas de dos modelos nos dice, en términos de probabilidades, qué tanto se ajusta mejor un modelo a los datos respecto al otro.**

## Factor de Bayes:

$B_{12} = \ln(\frac{P(D|H_1)}{P(D|H_2)}) = \ln(Z_1) - ln(Z_2)$ 

### Escala de Jeffreys:

| B | Fuerza de la evidencia a favor de $H_1$ |
|----|-----|
| $<1$ | muy escasa|
|1 a 2.5 | Significante|
|2.5 a 5 | Fuerte|
|$>5$ | Decisiva|
                

Ejemplo:

 Datos generados con: $y = -3.5x^2+3.6x-0.1$
    
   - Modelo 1: $y = ax^2+(-a+b)x+c$.
   - Modelo 2: $y = a\sin(bx)+c$
   
<img src="modelcomparisonplot.png">

Al usar un sampler anidado y calcular la evidencia bayesiana para ambos datos:

$B = log(\frac{Z_1}{Z_2})$
        
$\; \; \;= log(Z_1) - log(Z_2)$
        
$\; \; \; \;= -62.002 - (-255.064) $
        
$ \; \; \; \; = 193.062 $
    
$\; \; \; \; \; \; \implies$ Ventaja significativa para el modelo 1.

2. Se pueden definir el atributo value de los objetos Parameter según los resultados del optimizador, así mismo sus errores. Por ejemplo:
 - Ok_par = Parameter("Ok", -0.1, 0.01, (-0.2, 0.01), "\Omega_k")
