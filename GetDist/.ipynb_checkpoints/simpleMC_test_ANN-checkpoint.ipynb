```python
import matplotlib.pyplot as plt
from getdist import plots, MCSamples, chains
from getdist import *

import pylab

import numpy as np
import matplotlib
import tkinter
import pandas as pd

matplotlib.use('TkAgg')

%matplotlib inline
```

```python
dir_name = '/home/isidro/SimpleMC/chains'

roots = ['owaCDM_phy_HD+SN+BBAO+Planck_nested_dynesty_multi', 'owaCDM_phy_HD+SN+BBAO+Planck_nested_dynesty_multi_ANN']
```

```python
g= plots.getSinglePlotter(chain_dir = dir_name, width_inch= 10,
                          ratio = 0.9, 
                          analysis_settings={'smooth_scale_2D': 0.4,
                                             'smooth_scale_1D': 0.4, 
                                             'ignore_rows': 0.2})
```

```python
g.settings.axes_fontsize = 5
g.settings.lab_fontsize = 8
g.settings.legend_fontsize = 8
g.settings.axes_fontsize = 8

g.triangle_plot(roots, ['Om', 'Obh2', 'h', 'w', 'wa', 'Ok'], 
                diag1d_kwargs={'colors':['red','blue']},
                colors=['red','blue', 'green'],
                legend_labels=['dynesty only 50 live points', 'bambi 2 hidden layers, 200 neurons'], 
                filled = False, 
                normalized=True, shaded=False)
        
plt.savefig(dir_name+'owaCDMBambi.png', dpi=300)
plt.show()
```
