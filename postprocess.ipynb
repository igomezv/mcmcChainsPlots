```python
import numpy as np
import sys

class postprocess:
    def __init__ (self, filename):
        self.filename = filename
        
    def covarianceDiag(self):
        errors = np.loadtxt(self.filename)
        l = len(errors)
        if np.shape(errors) == (l,):
            return np.diag(errors**2)
        else:
            print("You need a single column file with the errors!")
            sys.exit(1)
    
    
        
```
